## Data Cleaning and Preprocessing

Cleaning is required to convert data into usable formats for analysis.


```python
df_copy = df.copy()
df_copy
```

```python
#reseting indexes 
df_copy.reset_index(drop=True , inplace=True)
```

```python
df_copy
```

```python
df_copy[~df_copy.Reviews.str.isnumeric()]
```

```python
df_copy = df_copy.drop(df_copy.index[9990])
```

```python
df_copy['Reviews'] = df_copy['Reviews'].astype(int)
df_copy['Reviews'].dtype
```

```python
df['Size'].unique()
```

```python
#1MB = 1024kb 

def size_processing(item):
    if str(item)[-1]=='M':
        res = float(str(item).replace('M',''))
        res*1024
        return res
    elif str(item)[-1] == 'k':
        res = float(str(item).replace('k',''))
        return res
    else:
        return str(np.nan)
```

```python
df_copy['Size'] = df_copy['Size'].apply(size_processing)
```

```python
df_copy.Size = df_copy.Size.astype("float")
```

```python
df_copy.Size.dtype
```

```python
df_copy.info()
```

```python
df_copy['Installs']
```

```python
df_copy['Installs'] = df_copy['Installs'].str.replace('+','').str.replace(',','')
```

```python
  # df_copy['Installs'] = df_copy['Installs'].astype("int")
```

```python
char_to_convert = ['+',',','$']
col_to_remove = ['Installs','Price']

for char in char_to_convert:
    for col in col_to_remove:
        df_copy[col] = df_copy[col].str.replace(char , '')
```

```python
df_copy['Price'].unique()
```

```python
df_copy['Price'] = df_copy['Price'].astype(float)
df_copy['Installs'] = df_copy['Installs'].astype(int)
```

```python
df_copy.info()
```

```python
df_copy['Last Updated']
```

```python
df_copy['Last Updated'] = pd.to_datetime(df_copy['Last Updated'])
```

```python
df_copy['Last Updated']
```

```python
df_copy['day'] = df_copy['Last Updated'].dt.day
df_copy['month'] = df_copy['Last Updated'].dt.month
df_copy['year'] = df_copy['Last Updated'].dt.year
```

```python
df_copy
```

```python
df_copy.drop("Last Updated" , axis=1 , inplace=True)
```

```python
df_copy
```

## Distribution of Free and Paid Applications

This count plot shows the number of free and paid applications available on the Google Play Store.  
It helps in understanding the dominance of app type in the market.


```python
plt.figure(figsize=(6,4))
sns.countplot(x=df_copy['Type'])
plt.title("Distribution of Free vs Paid Apps")
plt.xlabel("App Type")
plt.ylabel("Number of Apps")
plt.show()
```

### Insight
- Free applications dominate the Play Store
- Paid apps form a much smaller portion of the total apps
- Developers prefer free apps to reach a larger audience


## Content Rating Distribution

This plot shows how applications are categorized based on content suitability.


```python
plt.figure(figsize=(8,4))
sns.countplot(x=df_copy['Content Rating'])
plt.title("Distribution of Apps by Content Rating")
plt.xlabel("Content Rating")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

### Insight
- Majority of apps are rated for Everyone
- Very few apps target mature audiences
- Play Store prioritizes broad user accessibility


## Price Distribution of Applications

Markdown
This distribution plot visualizes the spread of application prices on the Google Play Store.
It helps understand how frequently apps are priced at different levels.


```python
#for the numeric columns
plt.figure(figsize=(8,5))
sns.distplot(df_copy['Price'], kde=True)
plt.title("Distribution of App Prices")
plt.xlabel("Price")
plt.ylabel("Density")
plt.show()
```

### Insight
- Most applications are priced at **0**, indicating a dominance of free apps
- Paid apps are relatively rare
- Only a small number of apps have higher prices


## Distribution of App Ratings

This plot shows the distribution of user ratings for applications on the Google Play Store.  
It helps in understanding overall user satisfaction and rating trends.


```python
plt.figure(figsize=(8,5))
sns.displot(df_copy['Rating'], bins=20)
plt.title("Distribution of App Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

### Insight
- Most applications have ratings between **4.0 and 4.5**
- Very few apps have extremely low ratings
- Indicates generally positive user feedback across the Play Store


## Kernel Density Estimation (KDE) of App Ratings

This KDE plot visualizes the probability density of app ratings on the Google Play Store.  
It helps understand where ratings are most concentrated and the overall distribution shape.


```python
plt.figure(figsize=(8,5))
sns.kdeplot(df_copy['Rating'], fill=True)
plt.xlabel("Rating")
plt.ylabel("Density")
plt.title("Density Distribution of App Ratings")
plt.show()
```

### Insight
- Ratings are highly concentrated between **4.0 and 4.5**
- The distribution is slightly left-skewed
- Most apps receive positive user feedback


## Distribution of App Type and Content Rating

This visualization shows the distribution of applications based on:
- **Type** (Free vs Paid)
- **Content Rating** (target audience suitability)

The count plots help understand how apps are categorized in terms of pricing model and audience restrictions on the Google Play Store.


```python
plt.figure(figsize=(10,10))
category = ['Type' , 'Content Rating']
for i in range(0,len(category)):
    plt.subplot(2,2,i+1)
    sns.countplot(x = df_copy[category[i]],palette='Set2')
    plt.xlabel(category[i])
    plt.xticks(rotation = 45)
    plt.tight_layout()
```

### Insight
- **Free apps** dominate the Play Store compared to paid apps
- Most applications are rated **"Everyone"**, indicating a focus on general audiences
- Apps with stricter content ratings form a smaller portion of the dataset
- Developers primarily target a broad user base with free, family-friendly applications


## App Distribution Across Categories

This pie chart represents the proportion of applications available in each category on the Google Play Store.  
It helps in understanding which categories have the highest number of apps.


```python
category_count = df_copy['Category'].value_counts()

plt.figure(figsize=(15,15))
plt.pie(category_count ,labels=category_count.index , autopct='%1.1f%%' , shadow=True)
plt.title('Which category is famous')
plt.show()
```

### Insight
- The **Family** and **Game** categories contain the highest number of apps
- This indicates high competition in these categories
- Niche categories have comparatively fewer applications


## Distribution of Application Sizes

This histogram represents the distribution of application sizes available on the Google Play Store.
Analyzing app size helps understand storage requirements and how app size may impact user downloads,
especially in regions with limited storage or network constraints.


```python
plt.figure(figsize=(8,8))
sns.histplot(df_copy['Size'] , bins = 30)

plt.title('Distribution of App Sizes on Google Play Store')
plt.xlabel('App Size (MB)')
plt.ylabel('Number of Apps')

plt.show()
```

### Insight
- Most applications have relatively small file sizes
- Very large-sized apps are less common
- Smaller app sizes are preferred as they reduce storage and data usage for users


```python
install_category = df_copy.groupby(["Category"])["Installs"].sum().sort_values(ascending = False).reset_index()
```

## Total Installs by App Category

This bar chart visualizes the total number of installs aggregated for each application category on the Google Play Store.

It helps identify which categories attract the highest number of users and reflects overall user demand across different app domains.


```python
plt.figure(figsize=(10,10))
sns.barplot(data=install_category , x = 'Category' , y='Installs')
plt.title("Total Installs by App Category")
plt.xlabel("Category")
plt.xticks(rotation = 90)
plt.ylabel("Total Install")
plt.show()
```

## Distribution of App Ratings (Box Plot)

This box plot visualizes the distribution of app ratings on the Google Play Store.  
It helps identify the median rating, spread of ratings, and presence of outliers.


```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df_copy["Rating"])
plt.title("Distribution of App Ratings")
plt.xlabel("Rating")
plt.show()
```

### Insight
- The median app rating lies around **4.3**
- Most applications have ratings between **4.0 and 4.6**
- Very few low-rated apps exist, indicating generally positive user feedback
- Outliers suggest some apps receive exceptionally low or high ratings


## Relationship Between Installs and Ratings

This boxplot examines how app ratings vary across different install levels.


```python
plt.figure(figsize=(10,10))
sns.boxplot( x="Installs", y="Rating", data=df_copy)
plt.title("Relationship Between Installs and Ratings")
plt.xlabel("Number of Installs")
plt.xticks(rotation = 90)
plt.ylabel("App Rating")

plt.show()
```

### Insight
- High install apps generally maintain good ratings
- Rating variation is higher for low-install apps
- Popular apps tend to be more stable in quality
